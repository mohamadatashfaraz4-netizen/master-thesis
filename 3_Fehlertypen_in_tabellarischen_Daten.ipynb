```python
from google.colab import files
uploaded = files.upload()

!pip install -q xlrd

import pandas as pd
import numpy as np

df = pd.read_excel('credit_card_corrupted_all_errors.xlsx')

# 1. Fehlende Werte
missing_per_column = df.isna().sum()
print('Missing values:')
print(missing_per_column)

# 2. Dubletten
duplicate_rows = df[df.duplicated()]
print('Number of duplicate rows:', len(duplicate_rows))

# 3. Mögliche Ausreißer über IQR-Regel für numerische Spalten
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(col, 'outliers:', len(outliers))

# 4. Beispielhafte Prüfung auf negative Werte in Betragsfeldern
amount_columns = [
    col for col in df.columns
    if 'amount' in col.lower()
    or 'revenue' in col.lower()
    or 'bill' in col.lower()
    or 'pay_amt' in col.lower()
]

for col in amount_columns:
    invalid_negative = df[df[col] < 0]
    print(col, 'negative values:', len(invalid_negative))
```
