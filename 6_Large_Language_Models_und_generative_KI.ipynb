```python
import pandas as pd

# Excel-Datensatz laden
file_name = "credit_card_corrupted_all_errors.xlsx"
df = pd.read_excel(file_name)

# Spaltennamen standardisieren
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

# Ein kompaktes Datenprofil erstellen
profile = pd.DataFrame({
    "dtype": df.dtypes.astype(str),
    "missing_count": df.isna().sum(),
    "missing_ratio": df.isna().mean(),
    "unique_values": df.nunique()
})

# Nur eine kleine Stichprobe verwenden, um unnötige Datenübertragung zu vermeiden
sample_rows = df.head(5).to_dict(orient="records")

prompt = f"""
Sie sind ein Datenqualitätsanalyst.
Basierend auf dem folgenden Datenprofil und den Beispielzeilen schlagen Sie mögliche Regeln zur Datenbereinigung vor.
Geben Sie die Antwort als nummerierte Liste zurück und unterscheiden Sie zwischen sicheren Regeln und Regeln, die eine fachliche Validierung erfordern.

Datenprofil:
{profile.to_string()}

Beispielzeilen:
{sample_rows}
"""

print(prompt)
```

```python
import os
from getpass import getpass

os.environ["OPENAI_API_KEY"] = getpass("Geben Sie Ihren OpenAI-API-Schlüssel ein: ")
```

```python
from openai import OpenAI

client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

model_name = "gpt-4.1-mini"

response = client.responses.create(
    model=model_name,
    input=prompt
)

print(response.output_text)
```

```python
with open("chapter6_llm_prompt.txt", "w", encoding="utf-8") as f:
    f.write(prompt)

from google.colab import files
files.download("chapter6_llm_prompt.txt")
```

```python
llm_output = response.output_text

with open("chapter6_llm_output.txt", "w", encoding="utf-8") as f:
    f.write(llm_output)

from google.colab import files
files.download("chapter6_llm_output.txt")
```
