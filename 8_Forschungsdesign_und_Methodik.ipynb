```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
# Originalen Credit-Card-Datensatz laden
credit_original = pd.read_excel(
    "default of credit card clients.xls",
    header=1
)

# Spaltennamen standardisieren
credit_original.columns = (
    credit_original.columns
      .astype(str)
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

print("Original Credit-Card-Dataset:")
print("Shape:", credit_original.shape)
print("Missing values:", int(credit_original.isna().sum().sum()))
print("Duplicates:", int(credit_original.duplicated().sum()))

credit_original.head()
```

```python
# Künstlich beschädigten Credit-Card-Datensatz laden
credit_corrupted = pd.read_excel(
    "credit_card_corrupted_all_errors.xlsx"
)

# Spaltennamen standardisieren
credit_corrupted.columns = (
    credit_corrupted.columns
      .astype(str)
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

print("Corrupted Credit-Card-Dataset:")
print("Shape:", credit_corrupted.shape)
print("Missing values:", int(credit_corrupted.isna().sum().sum()))
print("Duplicates:", int(credit_corrupted.duplicated().sum()))

credit_corrupted.head()
```

```python
# Vergleich zwischen Originaldaten und beschädigter Kopie
dataset_comparison = pd.DataFrame({
    "dataset": [
        "credit_card_original",
        "credit_card_corrupted_all_errors"
    ],
    "rows": [
        len(credit_original),
        len(credit_corrupted)
    ],
    "columns": [
        credit_original.shape[1],
        credit_corrupted.shape[1]
    ],
    "missing_values_total": [
        int(credit_original.isna().sum().sum()),
        int(credit_corrupted.isna().sum().sum())
    ],
    "duplicates": [
        int(credit_original.duplicated().sum()),
        int(credit_corrupted.duplicated().sum())
    ]
})

dataset_comparison
```

```python
# Breast-Cancer-Datensatz aus scikit-learn laden
breast_data = load_breast_cancer()

breast_df = pd.DataFrame(
    breast_data.data,
    columns=breast_data.feature_names
)

breast_df["target"] = breast_data.target

print("Breast-Cancer-Dataset:")
print("Shape:", breast_df.shape)
print("Missing values:", int(breast_df.isna().sum().sum()))
print("Duplicates:", int(breast_df.duplicated().sum()))

breast_df.head()
```

```python
def inject_missing_values(df, rate, random_state=42, exclude_columns=None):
    """Fehlende Werte zellweise in eine Kopie des DataFrames einfügen."""
    rng = np.random.default_rng(random_state)
    corrupted = df.copy()

    if exclude_columns is None:
        exclude_columns = []

    mask = rng.random(corrupted.shape) < rate

    for col in exclude_columns:
        if col in corrupted.columns:
            col_index = corrupted.columns.get_loc(col)
            mask[:, col_index] = False

    corrupted = corrupted.mask(mask)
    return corrupted


def inject_outliers(df, rate, factor=10, random_state=42, exclude_columns=None):
    """Numerische Ausreißer in ausgewählten Zeilen erzeugen."""
    rng = np.random.default_rng(random_state)
    corrupted = df.copy()

    if exclude_columns is None:
        exclude_columns = []

    numeric_cols = [
        col for col in corrupted.select_dtypes(include="number").columns
        if col not in exclude_columns
    ]

    row_count = int(len(corrupted) * rate)

    if row_count > 0 and len(numeric_cols) > 0:
        row_indices = rng.choice(corrupted.index, size=row_count, replace=False)
        corrupted.loc[row_indices, numeric_cols] = (
            corrupted.loc[row_indices, numeric_cols] * factor
        )

    return corrupted


def inject_duplicates(df, rate, random_state=42):
    """Dubletten durch Kopieren ausgewählter Zeilen erzeugen."""
    rng = np.random.default_rng(random_state)
    duplicate_count = int(len(df) * rate)

    if duplicate_count <= 0:
        return df.copy()

    duplicate_indices = rng.choice(df.index, size=duplicate_count, replace=False)
    duplicates = df.loc[duplicate_indices].copy()

    corrupted = pd.concat([df, duplicates], ignore_index=True)
    return corrupted


def inject_noisy_labels(df, target_col, rate, random_state=42):
    """Fehlerhafte Labels in einer binären Zielspalte erzeugen."""
    rng = np.random.default_rng(random_state)
    corrupted = df.copy()

    valid_indices = corrupted[corrupted[target_col].isin([0, 1])].index
    noisy_count = int(len(valid_indices) * rate)

    if noisy_count > 0:
        noisy_indices = rng.choice(valid_indices, size=noisy_count, replace=False)
        corrupted.loc[noisy_indices, target_col] = (
            1 - corrupted.loc[noisy_indices, target_col]
        )

    return corrupted
```

```python
# Zielspalte des Breast-Cancer-Datensatzes
breast_target_col = "target"

# Künstliche Fehlerszenarien erstellen
breast_missing = inject_missing_values(
    breast_df,
    rate=0.05,
    random_state=42,
    exclude_columns=[breast_target_col]
)

breast_outliers = inject_outliers(
    breast_df,
    rate=0.02,
    factor=10,
    random_state=42,
    exclude_columns=[breast_target_col]
)

breast_duplicates = inject_duplicates(
    breast_df,
    rate=0.01,
    random_state=42
)

breast_noisy_labels = inject_noisy_labels(
    breast_df,
    target_col=breast_target_col,
    rate=0.03,
    random_state=42
)

print("Breast missing values:", int(breast_missing.isna().sum().sum()))
print("Breast outliers shape:", breast_outliers.shape)
print("Breast duplicates:", int(breast_duplicates.duplicated().sum()))
print("Breast noisy labels target distribution:")
print(breast_noisy_labels[breast_target_col].value_counts())
```

```python
# Merkmale und Zielvariable trennen
X_breast = breast_df.drop(columns=[breast_target_col])
y_breast = breast_df[breast_target_col]

# Reproduzierbaren Train-Test-Split erstellen
X_train_breast, X_test_breast, y_train_breast, y_test_breast = train_test_split(
    X_breast,
    y_breast,
    test_size=0.33,
    random_state=42,
    stratify=y_breast
)

split_report = pd.DataFrame({
    "dataset": ["breast_cancer"],
    "train_rows": [len(X_train_breast)],
    "test_rows": [len(X_test_breast)],
    "features": [X_breast.shape[1]],
    "target_classes": [y_breast.nunique()]
})

split_report
```
